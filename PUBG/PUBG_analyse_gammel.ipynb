```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv

#Henter data:
reader = csv.reader(open("PUBGdata_rå.csv"), delimiter=";")
data_med_navne = pd.DataFrame(reader)

#Sletter navnene på variablene for at få rådatatal:
data = data_med_navne.drop(0,0)


data[0] = data[0].astype(int)
data[1] = data[1].astype(int)
data[2] = data[2].astype(int)
data[3] = data[3].astype(int)
data[4] = data[4].astype(int)
#Der er strings 'anton' og 'kenneth' i kolonne 5, så det laves ikke om til int.



#Definerer data:
spil = data[0]
kills_total = data[1]
antal_spillere = data[2]
plads = data[3]
dag = data[4]
ekstra_spiller = data[5]

#Definerer navne på kolonnerne:
data.columns = ['spil','kills_total','antal_spillere','plads','dag','ekstra_spiller']








```

```python
#Tæller chicken dinners:
chicken = 0
andenplads = 0
tredjeplads = 0
for element in plads: 
    if element==1: 
        chicken+=1
for element in plads:
    if element==2:
        andenplads+=1
for element in plads:
    if element==3:
        tredjeplads+=1
print('Chicken Dinners: ' + str(chicken))
print('Andenpladser: ' + str(andenplads))
print('Tredjepladser: ' + str(tredjeplads))
```

```python
#Tegner data
kills = plt.scatter(spil,kills_total, s=10, label = 'Antal kills Total')
plt.xlabel('Spil')
plt.ylabel('Antal Kills')
plt.title('Antal kills i PUBG')
plt.legend(loc='upper left')
plt.show(kills)
```

```python
#Justerer antal kills til antal kills per mand
kills_per_mand = kills_total / antal_spillere

#Gennemsnit og standardafvigelse:
gns_killspermand = kills_per_mand.mean()
std_killspermand = kills_per_mand.std()
analyse = [gns_killspermand, std_killspermand]
print(analyse)


plt.scatter(spil,kills_total, s=10, label='Antal kills per mand')
plt.xlabel('Spil')
plt.ylabel('Antal Kills')
plt.title('Antal kills per mand i PUBG')
plt.legend(loc='upper left')
plt.savefig("john.pdf")
plt.show
```
